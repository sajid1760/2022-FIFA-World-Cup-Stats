```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://fbref.com/en/comps/1/keepers/World-Cup-Stats"
browser.visit(url)
```

```python
browser.quit()
```

```python
html = browser.html
soup = BeautifulSoup(html)
type(soup)
```

```python
table = soup.find("table", {"id": "stats_keeper"})
table
```

```python
dfs = pd.read_html(str(table))
df_goalies = dfs[0]
df_goalies
```

```python
df_goalies.columns
```

```python
df_keepers_csv = df_goalies.iloc[:,np.arange(1,26).tolist()]
```

```python
keeperscols = ['Player',
 'Position',
 'Squad',
 'Current age',
 'Club',
 'Year of birth',
 'Matches Played',
 'Starts',
 'Minutes',
 '90s Played',
 'Goals Against',
 'Goals Against/90',
 'Shots on Target Against',
 'Saves',
 'Save Percentage',
 'Wins',
 'Draws',
 'Losses',
 'Clean Sheets',
 'Clean Sheet Percentage',
 'Penalty Kicks Attempted',
 'Penalty Kicks Allowed',
 'Penalty Kicks Saved',
 'Penalty Kicks Missed',
 'Save% (Penalty Kicks)']
```

```python
df_keepers_csv.columns = keeperscols
```

```python
for i in range(len(df_keepers_csv)):
    if (i % 26 == 25):
        df_keepers_csv = df_keepers_csv.drop(i)
df_keepers_csv = df_keepers_csv.reset_index(drop=True)
```

```python
df_keepers_csv.head()
```

```python
df_keepers_csv.to_csv("keepers.csv", index=False)
```
